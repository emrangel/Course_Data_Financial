### El método `.info`

El método `.info` es una de las herramientas más potentes para el análisis de datos, ya que devuelve un diccionario de Python con metadatos y métricas fundamentales extraídas directamente de Yahoo Finance.

Sin embargo, debes tener en cuenta que los campos disponibles **varían según el tipo de instrumento** (una acción tiene `forwardPE`, pero un ETF o una criptomoneda no).

A continuación las categorías principales de características (features) que puedes extraer:

**1. Métricas de Identificación y Perfil**
Son datos básicos para categorizar el activo en tu análisis de clustering.

* `shortName` / `longName`: Nombre oficial del fondo o empresa.

* `quoteType`: Tipo de activo (EQUITY, ETF, CRYPTOCURRENCY).

* `symbol`: El ticker (ej. "VOO").

* `longBusinessSummary`: Descripción textual (útil si quieres hacer procesamiento de lenguaje natural o NLP para agrupar por temas).

**2. Métricas de Riesgo y Rendimiento (Claves para Clustering)**
Estas son las más importantes para tu modelo matemático:

* `beta` / `beta3Year`: Mide la volatilidad relativa al mercado. Fundamental para separar activos refugio de activos de riesgo.

* `yield` / `dividendYield`: El rendimiento por dividendo. Clave para identificar clusters de "Income" (ingresos).

* `fiveYearAvgDividendYield`: Estabilidad de los dividendos a largo plazo.

* `trailingAnnualReturn`: Rendimiento histórico reciente.

* `morningStarRiskRating`: (Solo ETFs) Calificación de riesgo de Morningstar.

**3. Datos de Costes y Tamaño**
Para comparar la eficiencia y liquidez de los instrumentos:

* `annualReportExpenseRatio`: El ratio de gastos (comisiones). Diferencia claramente los fondos de gestión pasiva de los de gestión activa.

* `totalAssets`: El patrimonio bajo gestión (AUM). Indica el tamaño y la relevancia del fondo.

* `averageVolume`: Volumen medio de negociación (liquidez).

* `marketCap`: Capitalización bursátil (para acciones).

**4. Estadísticas de Precio (Acción del Precio)**
* `fiftyTwoWeekHigh` / `fiftyTwoWeekLow`: Máximos y mínimos anuales.

* `fiftyDayAverage` / `twoHundredDayAverage`: Medias móviles, útiles para detectar tendencias técnicas.

* `52WeekChange`: El cambio porcentual del precio en el último año.

**Ejemplo de implementacion de** `.info`

````python
# Definir tiker o lista de tickers
tickers = ["AAPL", "MSFT", "GOOGL"]
````

````python
# Crear función para extraer múltiples características y convertirlas en DataFrame


def get_etf_data(ticker_list):
    all_features = []

    for symbol in ticker_list:
        etf = yf.Ticker(symbol)
        info = etf.info
        # Extraer algunas caracteristicas
        # Nota: revisa las características disponibles según el ticker
        features = {
            "Ticker": symbol,
            "Name": info.get("shortName"),
            "Category": info.get("category"),
            "Total Assets": info.get("totalAssets"),
            "Yield": info.get("yield"),
            "Expense Ratio": info.get("annualReportExpenseRatio"),
            "Beta": info.get("beta3Year"),
            "52WeekChange": info.get("52WeekChange"),
            "Three Year Return": info.get("threeYearAverageReturn"),
            "Morningstar Risk": info.get("morningStarRiskRating"),
        }
        all_features.append(features)

    return pd.DataFrame(all_features)
````

````python
# crear DataFrame con múltiples tickers
df_tickers = get_etf_data(tickers)
df_tickers
````

### **Ejemplo: Análisis de Volatilidad del S&P500**

leeremos la lista de empresas directamente desde Wikipedia usando Pandas y luego le pediremos a yfinance que nos traiga sus precios.

````python
# 1. Leemos la tabla de Wikipedia (la función read_html devuelve una lista de tablas)
import requests

url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
response = requests.get(url, headers=headers)
tablas = pd.read_html(response.text)

# La primera tabla [0] contiene los componentes actuales
df_sp500 = tablas[0]

# Extraemos solo la columna 'Symbol' (los tickers)
tickers = df_sp500['Symbol'].tolist()

# Yahoo Finance usa '-' en lugar de '.' para algunos tickers (ej. BRK.B -> BRK-B)
tickers = [t.replace('.', '-') for t in tickers]

print(f"Se han encontrado {len(tickers)} activos.")
````

**Calculo de la Volatilidad**

Para calcular la volatilidad, necesitamos los precios de cierre diarios.

````python
# 2. Descargamos solo el precio de cierre ('Close') del último mes
# Usamos threads=True para que sea mucho más rápido
datos = yf.download(tickers, period="1mo")['Close']

# Eliminamos columnas que tengan muchos datos faltantes (opcional, por limpieza)
datos = datos.dropna(axis=1, thresh=15)
````

En finanzas, la **volatilidad** suele medirse como la **desviación estándar de los retornos diarios.**

**Fórmula: **
1. Calculamos el cambio porcentual diario.
2. Calculamos la desviación estándar de esos cambios.

````python
# 3. Calcular retornos diarios (cambio porcentual entre el día actual y el anterior)
retornos = datos.pct_change()

# 4. Calcular la volatilidad (Desviación Estándar) de cada columna (acción)
volatilidad = retornos.std()

# Convertimos el resultado en un ranking ordenado de mayor a menor
ranking_volatilidad = volatilidad.sort_values(ascending=False)

# Mostramos los 5 más volátiles
print("--- Las 5 acciones más volátiles del mes ---")
print(ranking_volatilidad.head(5))
````

uando ejecutas este código, obtienes un número pequeño (por ejemplo, 0.035).

**¿Qué significa?** Es la desviación típica diaria. Si multiplicas ese número por la raíz cuadrada de 252 (días hábiles del año), obtendrás la **volatilidad anualizada**, que es el estándar de la industria.

````python
# Volatilidad Anualizada del top 1
vol_anual = ranking_volatilidad.iloc[0] * np.sqrt(252)
print(f"\nLa acción más volátil tiene una volatilidad anualizada de: {vol_anual:.2%}")
````

### Ejemplo: Comparativa de Margen de Utilidad

El Margen de Utilidad (Net Profit Margin) nos dice cuántos centavos de ganancia neta genera una empresa por cada dólar de ingresos.

Para este ejercicio, compararemos a tres gigantes tecnológicos: **Apple** (AAPL), **Microsoft** (MSFT) y **Google** (GOOG).

````python
# 1. Definimos los tickers de las empresas a comparar
tickers = ["AAPL", "MSFT", "GOOG"]
margenes = {}

for t in tickers:
    # Creamos el objeto Ticker para cada una
    empresa = yf.Ticker(t)

    # Extraemos el Estado de Resultados (Income Statement)
    # .financials devuelve los datos anuales
    estado_resultados = empresa.financials

    # Extraemos los datos más recientes (columna 0)
    # 'Net Income' es la Utilidad Neta
    # 'Total Revenue' son los Ingresos Totales
    utilidad_neta = estado_resultados.loc['Net Income'].iloc[0]
    ingresos_totales = estado_resultados.loc['Total Revenue'].iloc[0]

    # Calculamos el Margen de Utilidad (Net Margin)
    margen = (utilidad_neta / ingresos_totales) * 100

    # Guardamos el resultado en nuestro diccionario
    margenes[t] = margen

# 2. Convertimos los resultados en una tabla para visualizar mejor
df_comparativo = pd.DataFrame.from_dict(margenes, orient='index', columns=['Margen de Utilidad (%)'])

# 3. Mostramos el resultado ordenado de mayor a menor
print(df_comparativo.sort_values(by='Margen de Utilidad (%)', ascending=False))
````

**Explicación del Código y Resultado**

`empresa.financials`: Este comando nos trae una tabla donde las filas son conceptos contables. Usamos `.loc['Net Income']` para "saltar" directamente a la fila de Utilidad Neta.

`.iloc[0]`: Los estados financieros en `yfinance` suelen estar ordenados de la fecha más reciente a la más antigua. Al usar `[0]`, nos aseguramos de tomar el último año fiscal reportado.

**La lógica del Margen**: Un margen del **25%** significa que, tras pagar todos los costos, impuestos y gastos, la empresa se queda con $0.25 de cada dólar vendido.

**¿Por qué es útil esto para una persona de finanzas?**

**Eficiencia Operativa**: Si dos empresas venden lo mismo pero una tiene un margen del 30% y la otra del 15%, la primera es mucho más eficiente convirtiendo ventas en ganancias.

**Detección de Ventajas Competitivas**: Las empresas con "fosos económicos" (moats) suelen mantener márgenes altos y estables a lo largo del tiempo.

````python
df_comparativo.plot(kind='bar', color='skyblue', legend=False)
plt.title('Comparativa de Margen de Utilidad Neta')
plt.ylabel('Porcentaje (%)')
plt.show()
````
